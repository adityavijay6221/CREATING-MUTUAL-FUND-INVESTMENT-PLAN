```python
import pandas as pd
df=pd.read_csv('nifty50_closing_prices.csv')
df.head()
```

```python
df.info()
```

```python
#CHECKING NULL VALUES
df.isnull().sum()
```

```python
#FILLING MISSING VALUES
df.fillna(method='ffill',inplace=True)
df.info()
```

```python
import datetime as dt
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df.set_index('Date',inplace=True)
df.head()
```

```python
#VISUALIZE CLOSING PRICES OF ALL 50 companies
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
plt.plot(df)
plt.legend(df.columns)
plt.show()
```

```python
#TOP 10 most volatile companies
volatility=pd.DataFrame(df.std())
volatility.rename(columns={0:'Volatility'},inplace=True)
top_10_most_volatile=volatility['Volatility'].sort_values(ascending=False).head(10)
top_10_most_volatile
```

```python
#COMPANIES WITH BEST GROWTH RATE
growth_rate=df.pct_change()*100
avg_growth_rate=growth_rate.mean()
top_10_highest_growth_rate=avg_growth_rate.sort_values(ascending=False).head(10)
top_10_highest_growth_rate
```

```python
#COMAPNIES WITH HIGHEST ROI
initial_price=df.iloc[0]
initial_price
final_price=df.iloc[-1]
final_price
roi=(final_price-initial_price)/initial_price*100
top_10_highest_roi=roi.sort_values(ascending=False).head(10)
top_10_highest_roi
```

Creating a Mutual Fund Plan Based on High ROI and Low Risk

```python
#SETTING ROI AND VOLATILITY THRESHOLDS
roi_threshold=roi.median()
volatility_threshold=volatility['Volatility'].median()
selected_companies=df.columns[(roi>roi_threshold) & (volatility['Volatility']<volatility_threshold)]
selected_companies
```

```python
selected_volatilities=volatility.loc[selected_companies]
selected_volatilities
```

```python
selected_roi=roi.loc[selected_companies]
selected_roi
```

```python
#To balance the investment between these companies, we can use an inverse volatility ratio for allocation.
#Companies with lower volatility will get a higher weight. Let’s calculate the weight for each company.
selected_volatilities['Inverse Volatility']=1/selected_volatilities['Volatility']
selected_volatilities
```

```python
selected_volatilities['Weights']=(selected_volatilities['Inverse Volatility']/selected_volatilities['Inverse Volatility'].sum())*100
selected_volatilities
```

ANALYZING OUR MF COMPANIES

```python
#RISK COMPARISON(VOLATILITY): GROWTH COMPANIES VS MF COMPANIES
risk_mf_companies=volatility.loc[selected_companies]
risk_growth_companies=volatility.loc[top_10_highest_growth_rate.index]
plt.figure(figsize=(20,10))
sns.barplot(y=risk_mf_companies.index,x=risk_mf_companies['Volatility'], orient='h', label='MF Companies')
sns.barplot(y=risk_growth_companies.index,x=risk_growth_companies['Volatility'], orient='h', label='Growth Companies')
plt.title('Volatility Comparison: MF Companies vs Growth Companies')
plt.xlabel('Volatility')
plt.ylabel('Company')
plt.legend()
plt.show()
```

```python
#GROWTH COMAPRISON(ROI): GROWTH COMPANIES VS MF COMPANIES
growth_mf_companies=roi.loc[selected_companies]
growth_growth_companies=roi.loc[top_10_highest_growth_rate.index]
plt.figure(figsize=(20,10))
sns.barplot(y=growth_mf_companies.index,x=growth_mf_companies, orient='h', label='MF Companies')
sns.barplot(y=growth_growth_companies.index,x=growth_growth_companies, orient='h', label='Growth Companies')
plt.title('ROI Comparison: MF Companies vs Growth Companies')
plt.xlabel('ROI')
plt.ylabel('Company')
plt.legend()
plt.show()
```

CALCULATING EXPECTED RETURNS

```python
import numpy as np

monthly_investment = 5000  # Monthly investment in INR
years = [1, 3, 5, 10]  # Investment periods (in years)
n = 12  # Number of times interest is compounded per year (monthly)

avg_roi = growth_mf_companies.mean() / 100  # Convert to decimal

def future_value(P, r, n, t):
    return P * (((1 + r/n)**(n*t) - 1) / (r/n)) * (1 + r/n)

future_values = [future_value(monthly_investment, avg_roi, n, t) for t in years]
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=years, y=future_values)
plt.title('Future Value of Investment Over Time')
plt.xlabel('Years')
plt.ylabel('Future Value')
plt.show()
```
